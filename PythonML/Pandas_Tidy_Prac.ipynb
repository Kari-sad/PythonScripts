```python
## drop a column with na values
ref_dt = raw_csv_dt.dropna()
ref_dt
## As you can see from below it will wipe away all the data if there is a sinhle NA value in that row,, 
## in such cases it is better to drop the column with most NaN
```

```python
ref_dt = raw_csv_dt.dropna(how='all', axis=1) ## it drops the column 
## How takes 'all' and 'any' as arguments.
# all --> drops the column or row with all NaN values
# any --> drops the column or row with one or more NaN values
del(raw_csv_dt)
ref_dt
```

```python
## To delete the row with all NaN values
ref_dt = ref_dt.dropna(how='all', axis=0)
ref_dt
```

```python
## To fill NA/NaN data with filled with fillna(mrthod='')
# 'ffill' will replace the naN/Na values with what is there above
# 'bfil' will  replace the naN/Na values with what is there below.
ref_dt.fillna(method='ffill')
ref_dt.fillna(method='bfill')
```

```python
## To fill the values with mean
ref_dt.fillna(ref_dt.mean())
```

                               Grouping
Hirarchical Grouping
    1. DatFrame.groupby(['Col1' , 'Col2' ...]) --> devides the data in groups and creates a groupby object.
    2. DataFrame.ngroups() --> returns total number of groups DF is divided .
    3. DataFrame.size() --> returns the Total count of ROWS in each group
    4. DataFrame.count() -->returns the total number of non-NaN values in each column
    5. DataFrame.get_group('GroupName') --> To Check the content of specific group.
    6. DataFrame.groups --> returns all group names with the indices 
    7. DataFrame.set_index(['COL1' , 'COL2']) -->     Alternative way to create a group

```python
# grouping with single column classes
ref_grp_cname = ref_dt.groupby('CNAME')
ref_grp_cname
```

```python
## To check the number of groups
ref_grp_cname.ngroups
```

```python
## Check the count of rows in of each group
ref_grp_cname.size()
```

```python
## Displays the indexes which belong to the group bucket
ref_grp_cname.groups
```

```python
## To check the number of rows in each column
ref_grp_cname.count()
```

```python
## To check the content of a specific group
## KeyError: 'yahoo' if a group is not found then it will throw an error
ref_grp_cname.get_group('AMAZON').head()
```

```python
# To display the head of each of the group
ref_grp_cname.head(1)
```

```python
# To display the nth index in the group
ref_grp_cname.nth(0)
```
