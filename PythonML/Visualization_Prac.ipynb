# Digital Image Processing

```python
 def ImgPropChg(img):
    img = img
    nrows , ncols , ndim = img.shape
    row , col = np.ogrid[:nrows , :ncols]
    cnt_row , cnt_col = nrows/2 , ncols/2
    OuterMask = ((row - cnt_row)**2+(col - cnt_col)**2>(nrows/2)**2)
    img.setflags(write=1)
    img[OuterMask] = 0
    return img
```
